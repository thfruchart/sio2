#### Exercice 1 : total des valeurs d'un tableau

```python
t = [10,20,30,40,50]
i = 0
s = 0
while i < 5:
    v = t[i]
    s = s + v
    i = i+1

print('la somme des valeurs du tableau est : ')
print( s )
```

#### Exercice 2 : total des valeurs d'un tableau de longueur quelconque

```python
t = [10,20,30,40,50,60,70]
i = 0
s = 0
while i < len(t):  # seule cette ligne est à modifier
    v = t[i]
    s = s + v
    i = i+1

print('la somme des valeurs du tableau est : ')
print( s )
```

#### Exercice 3 : moyenne des valeurs d'un tableau

```python
t = [10,20,30,40,50,60,70]
i = 0
s = 0
while i < len(t): 
    v = t[i]
    s = s + v
    i = i+1

m = s/ len(t)    # seule cette ligne est à àjouter
print('la somme des valeurs du tableau est : ')
print( m )
```

#### Exercice 4 : nombre d'occurrences d'une valeur dans un tableau

```python
t = [10,20,30,20,30,20]
print(t)
v = int(input('entrer un entier : '))

nb = 0  # cette variable sert à compter le nombre de fois où la valeur v est dans le tableau t

i = 0
while i < len(t):     if v == t[i]: # si v est présente dans le tableau à l'indice i
        nb = nb+1 # alors le compteur nb est incrémenté
    i = i+1 #on passe à l'indice suivant

print('nombre de fois où cette valeur est dans t')
print( nb )
```

#### Exercice 5 : test si un tableau est trié

```python
t = [10,20,30,40,50]
resu = True  # tant qu'aucun problème n'est détécté, on considère que le tableau est bien trié
i = 0
while i < len(t)-1: # on parcourt les indices du tableau jusqu'à l'AVANT DERNIER
    if t[i] > t[i+1]: # si deux valeurs sont mal triées
        resu = False # alors le tableau n'est pas trié
    i = i+1 

if resu :
    print("le tableau est trié")
else :
    print("le tableau n'est pas trié")
```
